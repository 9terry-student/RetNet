```python
import numpy as np

def compute_advantage(rewards, values, gamma=0.99):
  advantages=[]
  # rewards: 실제 받은 보상
  # values: Critic이 예측한 기대 보상
  # advantage = 실제 - 기대
  for i in range(len(rewards)-1):
    advantages.append(rewards[i]+gamma*values[i+1]-values[i])
  advantages.append(rewards[len(rewards)-1]-values[len(rewards)-1])

  output=np.array(advantages)
  return output

def ppo_loss(old_probs, new_probs, A, e=0.2):
  L=[]
  # r_t = new_probs / old_probs
  for i in range(len(old_probs)):
    r_t=new_probs[i]/old_probs[i]

  # L = min(r_t * A, clip(r_t, 1-e, 1+e) * A)
    L.append(-np.minimum(r_t*A[i],np.clip(r_t,1-e,1+e)*A[i]))
  output=np.array(L)
  return output

def train_step(states, actions, rewards, values, old_probs, new_probs):
  # 1. Advantage 계산
  adv=compute_advantage(rewards,values)

  # 2. PPO loss 계산
  loss=ppo_loss(old_probs,new_probs,adv)

  # 3. 반환
  return loss
```

```python
# test
np.random.seed(42)
rewards  = np.array([1.0, 0.5, 1.0, 0.0])
values   = np.array([0.8, 0.6, 0.9, 0.1])
old_probs = np.array([0.3, 0.4, 0.2, 0.1])
new_probs = np.array([0.35, 0.38, 0.25, 0.12])

loss = train_step(None, None, rewards, values, old_probs, new_probs)
print("loss:", loss)
print("loss shape:", loss.shape)  # (4,)
```
